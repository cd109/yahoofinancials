# 丢弃包含缺失值的字段

```python
df.head() 
```

```python
df[df['建筑类别'].isnull()].head(3) #找出建筑类别为缺失值的数据
```

```python
df = df.dropna(axis=1, how='all')#全部为缺失值的列舍弃
```

```python
df.head(3)
```

# 填补缺失值

```python
df.head()
```

```python
df['物 业 费'] =df['物 业 费'].fillna(0)#物业费缺失值用0替换
```

```python
df.head()
```

```python
df['總價'].isnull().sum()#查看总价栏位的缺失值
```

```python
df[df['總價'].isnull()] #观察总价栏位缺失的数据
```

```python
(df['總價'] / df['建筑面积']).mean() #计算获取均价
```

```python
df['總價'] = df['總價'].fillna(df['建筑面积'] * (df['總價'] / df['建筑面积']).mean())#用平均值补齐总价的缺失值
```

```python
df['總價'].isnull().sum()#补齐后没缺失值了
```
